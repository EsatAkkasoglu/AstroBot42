```python
import feedparser
import json
import os

# Path to the JSON file where seen news items are stored
seen_news_file = 'seen_news.json'

# Load seen news items from the file if it exists, otherwise create an empty set
if os.path.exists(seen_news_file):
    with open(seen_news_file, 'r') as file:
        seen_news = set(json.load(file))
else:
    seen_news = set()
# RSS feed URLs of popular space news channels
rss_urls = {
    'NASA Breaking News': 'https://www.nasa.gov/rss/dyn/breaking_news.rss',
    'Space.com News': 'https://www.space.com/feeds/all',
    'Sky & Telescope - News': 'https://skyandtelescope.org/astronomy-news/feed/',
    'ESA Top News': 'https://www.esa.int/rssfeed/TopNews',
    'ESA Space Science News': 'https://www.esa.int/rssfeed/Our_Activities/Space_Science',
    'ESA Human Spaceflight and Exploration News': 'https://www.esa.int/rssfeed/HSF',
    'New Scientist - Space': 'https://www.newscientist.com/subject/space/feed/'
}

def fetch_news(rss_url):
    # Fetch and parse the RSS feed
    feed = feedparser.parse(rss_url)
    news_items = []
    
    for entry in feed.entries:
        # Extract the necessary information from each news item
        title = entry.title if 'title' in entry else 'No Title'
        link = entry.link if 'link' in entry else 'No Link'
        published = entry.published if 'published' in entry else 'No Publication Date'
        
        # Append the news item to the list
        news_items.append({
            'title': title,
            'link': link,
            'published': published
        })
    
    return news_items

# Dictionary to store news from all feeds
all_news = {}

for source, url in rss_urls.items():
    new_news_items = []
    news_items = fetch_news(url)
    for item in news_items:
        if item['link'] not in seen_news:  # Check if the news item is not in the seen news
            new_news_items.append(item)
            seen_news.add(item['link'])  # Add the new item to the seen news

    all_news[source] = new_news_items


# Print the news items in a formatted way
for source, news_items in all_news.items():
    print(f"Latest news from {source}:")
    for i, item in enumerate(news_items[:5], 1):  # Show only the top 5 news items from each source
        print(f"{i}. {item['title']}")
        print(f"   Link: {item['link']}")
        print(f"   Published: {item['published']}")
    print("\n")
with open(seen_news_file, 'w') as file:
    json.dump(list(seen_news), file)  # Convert the set to list as JSON doesn't support set data type
```

```python
import re
import sympy as sp  # LaTeX matematik ifadelerini çözmek için sympy kütüphanesini kullanıyoruz

def format_text(text):
    # Metni paragraflara bölmek için boş satırları kullanın
    paragraphs = text.split("\n\n")

    formatted_paragraphs = []
    for paragraph in paragraphs:
        # Satır başı düzenlemesi yapın (belirli bir karakter sınırına ulaşana kadar)
        lines = re.findall('.{1,70}(?:\s+|$)', paragraph)  # Her 70 karakterde bir bölün
        formatted_paragraph = "\n".join(lines)

        # LaTeX matematik ifadelerini çözün
        formatted_paragraph = sp.latex(formatted_paragraph, mode='inline')

        formatted_paragraphs.append(formatted_paragraph)

    return formatted_paragraphs

# Örnek metin
text = """
Summary: The presence of magnetic fields in the late inspiral of black hole -- neutron
star binaries could lead to potentially detectable electromagnetic precursor
transients. Using general-relativistic force-free electrodynamics simulations,
we investigate pre-merger interactions of the common magnetosphere of black
hole -- neutron star systems. We demonstrate that these systems can feature
copious electromagnetic flaring activity, which we find depends on the magnetic
field orientation but not on black hole spin. Due to interactions with the
surrounding magnetosphere, these flares could lead to Fast Radio Burst-like
transients and X-ray emission, with $\mathcal{L}_{\rm EM} \lesssim 10^{41}
\left( B_\ast/ 10^{12}\, \rm G \right)^2\, \rm erg/ s$ as an upper bound for
the luminosity, where $B_\ast$ is the magnetic field strength on the surface of
the neutron star."""

# Metni düzenle
formatted_text = format_text(text)

# Düzenlenmiş metni yazdır
for i, paragraph in enumerate(formatted_text, start=1):
    print(f"Paragraf {i}:\n{paragraph}\n")
```
